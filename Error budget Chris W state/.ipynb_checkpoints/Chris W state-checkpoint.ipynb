```python
import numpy as np
from qutip import*
import time
import matplotlib.pyplot as plt
import itertools as it
sqrt = np.sqrt
pi   = np.pi
class Add_gate:
    def __init__(gate, name, Tar_Con, angle):
        gate.name    = name
        gate.Tar_Con = Tar_Con
        gate.angle   = angle
                
Nqubits= 3
Nlevels= 3
Paulis_gt = 2.00e-2
CZ_gt = 0.353
gate_time_CCZS = CZ_gt/(sqrt(2))
gatetime_PI12 = 5.00e-2
gate_time_SWAP = gate_time_CCZS
Alpha = -2*pi*210
ZZ_li = [[0,1], [0,2]]
ZZ_stren = [2*pi*200e-3, 2*pi*200e-3]

def project_to_qubit(rho):
    N = int(len(rho.dims[0]))
    projector = Qobj(np.array([[1, 0, 0],[0, 1, 0]]))
    tot_proj = tensor([projector]*N)
    rho_proj = tot_proj*rho*tot_proj.dag()
    return rho_proj.unit()

T1s = [39.73, 27.35, 34.73]
T2s = [57.67, 45.21, 21.49]

Diss =T1s
Deph = []
for i in range(3):
    _1byT2 = 1/T2s[i]
    _1byT1 = 1/(2*T1s[i])
    Deph.append(1/(_1byT2 - _1byT1))

    
# Import the developed package
import Chalmers_Simulator_mesolve as CS
```

# Noisefree and in the absence of ZZ interaction

```python
# Create Hamiltonian and collapse operators
Hamiltonian, c_ops = CS.create_system_Hamiltonian(Nqubits, Nlevels, Paulis_gt, gatetime_PI12, CZ_gt, gate_time_CCZS, gate_time_SWAP, \
                                                  Alpha, Diss, Deph, Texc = [], ZZ_list=[], ZZ_strength=[])

Ini = tensor(basis(3,0), basis(3,0), basis(3,0))

gate=[]
gate.append(Add_gate(name=["PYe"   ], Tar_Con=[0],          angle=[2*np.arccos(1/sqrt(3))]        ))
gate.append(Add_gate(name=["PI12" ],  Tar_Con=[0],          angle=[pi]                            ))
gate.append(Add_gate(name=["SCCZS"],  Tar_Con=[[0,1,2,0]],  angle=[0]                             ))
gate.append(Add_gate(name=["PXe"  ],  Tar_Con=[0],          angle=[pi]                            ))


# Getting rid of the extra phase
Wstate = CS.Execute(Hamiltonian, [], gate, Ini)
```

```python
pzz = np.linspace(0,2*pi,25)
F = []
counter = 1
for phase in pzz:
    gate=[]
    gate.append(Add_gate(name=["PY"   ], Tar_Con=[0],         angle=[2*np.arccos(1/sqrt(3))]        ))
    gate.append(Add_gate(name=["PI12" ], Tar_Con=[0],         angle=[pi]                            ))
    gate.append(Add_gate(name=["SCCZS"], Tar_Con=[[0,1,2,0]], angle=[0]                             ))
    gate.append(Add_gate(name=["PX"   ], Tar_Con=[0],         angle=[pi]                            ))

    # Getting rid of the extra phase
    gate.append(Add_gate(name=["PZ"   ], Tar_Con=[2],         angle=[phase*pi]                      ))

    fstate = CS.Execute(Hamiltonian, [], gate, Ini)
    W001 = tensor(basis(2,0), basis(2,0), basis(2,1))
    W010 = tensor(basis(2,0), basis(2,1), basis(2,0))
    W100 = tensor(basis(2,1), basis(2,0), basis(2,0))
    Wstate = (W001 + W010 + W100).unit()
    F.append(fidelity(ket2dm(Wstate), project_to_qubit(ket2dm(fstate))))
    print(counter)
    counter = counter + 1
```
